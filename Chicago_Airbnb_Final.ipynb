### Research Question

What listing characteristics and neighborhood features determine nightly price for a short-term Airbnb rental in Chicago, and can we predict whether a listing will price above or below the city median?

### Motivation

Setting the right price is a hard decision for a new Airbnb host. Airbnb's own pricing suggestions are opaque, and hosts have no easy way to know which features of their listing actually justify a higher rate. This analysis builds a data-driven answer to that question by combining listing-level characteristics with neighborhood crime and income data to identify the true drivers of nightly price.

The target audience is **Airbnb hosts** looking for a data-backed pricing baseline and **property investors** evaluating Chicago neighborhoods for short-term rental returns.

### Approach

We combine four datasets across 6,162 active Chicago listings:

1. **Airbnb listing data** — room type, capacity, bathrooms, amenities, host characteristics, review scores
2. **Crime data** — non-domestic incident counts aggregated to community area via spatial join
3. **Boundary and community area data** — for spatial join with crime data
4. **Income data** — ACS median household income estimates for all 77 Chicago community areas

We run two model types across two feature sets (listing-only vs. listing + neighborhood) to isolate the sigificance of neighborhood context:
- **Linear regression** — predicts exact nightly price
- **Logistic regression** — classifies listings as above/below the $143 median

## Section 1 — Airbnb Listings Cleaning

The raw listings file from Inside Airbnb contains 75+ columns, many of which are URLs, scrape metadata, or duplicate fields with no analytical value. This section documents which columns were dropped and why, then handles type conversion, null imputation, and scope filtering.

**Source:** `listings.csv` — scraped September 22, 2025  


```python
airbnb = pd.read_csv('../data/raw/listings.csv')
print(f"Raw shape: {airbnb.shape}")
```

### 1.1 Column Drops

We drop columns in six categories:
- **URLs and scrape metadata** — no analytical value
- **Unstructured text fields** — unusable without NLP
- **Host identifiers** — personally identifiable, not predictive of price
- **Redundant listing counts** — covered by `calculated_host_listings_count`
- **Availability snapshot columns** — reflect calendar state at scrape time, not stable listing attributes
- **Leakage columns** — `estimated_revenue_l365d` is derived from price and would constitute data leakage if used as a predictor


```python
drop = [
    # URLs — no analytical value
    'listing_url', 'host_url', 'picture_url', 'host_thumbnail_url', 'host_picture_url',

    # Scrape metadata
    'scrape_id', 'source', 'calendar_updated', 'calendar_last_scraped', 'last_scraped',

    # Unstructured text — not usable without NLP
    'name', 'description', 'neighborhood_overview', 'host_about',

    # Host identifiers — personally identifiable
    'host_id', 'host_name', 'host_location', 'host_neighbourhood',
    'host_verifications', 'host_has_profile_pic', 'host_identity_verified',

    # Redundant listing counts
    'calculated_host_listings_count_entire_homes',
    'calculated_host_listings_count_private_rooms',
    'calculated_host_listings_count_shared_rooms',
    'host_total_listings_count', 'host_listings_count',

    # Host engagement — affect booking conversion, not price
    'host_response_rate', 'host_acceptance_rate',

    # Review subscores — keeping rating, cleanliness, location, value
    'reviews_per_month', 'review_scores_accuracy',
    'review_scores_checkin', 'review_scores_communication',

    # Availability snapshot
    'has_availability', 'availability_30', 'availability_60',
    'availability_90', 'availability_365', 'availability_eoy',

    # Minimum/maximum nights variants — minimum_nights covers this
    'minimum_minimum_nights', 'minimum_maximum_nights',
    'maximum_maximum_nights', 'minimum_nights_avg_ntm',
    'maximum_minimum_nights', 'maximum_nights_avg_ntm', 'maximum_nights',

    # Review timing
    'first_review', 'last_review',

    # Neighbourhood duplicates
    'neighbourhood', 'neighbourhood_group_cleansed',

    # Admin/legal
    'license',

    # Data leakage — revenue is derived from price
    'estimated_revenue_l365d', 'estimated_occupancy_l365d',

    # Booking activity — result of being listed, not a predictor of price
    'number_of_reviews_ltm', 'number_of_reviews_l30d', 'number_of_reviews_ly',

    # Text version of bathrooms — numeric column is cleaner
    'bathrooms_text',
]

airbnb = airbnb.drop(columns=drop)
print(f"Shape after column drops: {airbnb.shape}")
```

### 1.2 Type Conversion

```python
# Price: strip dollar signs and commas, cast to float
airbnb['price'] = (
    airbnb['price']
    .str.replace('[$,]', '', regex=True)
    .str.strip()
    .astype(float)
)

# Boolean flags: 't'/'f' strings to proper booleans
airbnb['instant_bookable']   = airbnb['instant_bookable'].str.strip().str.lower().map({'t': True, 'f': False}).astype('boolean')
airbnb['host_is_superhost']  = airbnb['host_is_superhost'].str.strip().str.lower().map({'t': True, 'f': False}).astype('boolean')

# Response time: ordered categorical
response_order = ['within an hour', 'within a few hours', 'within a day', 'a few days or more']
airbnb['host_response_time'] = pd.Categorical(airbnb['host_response_time'], categories=response_order, ordered=True)

# Host tenure: days since host joined as of scrape date
airbnb['host_since']        = pd.to_datetime(airbnb['host_since'], errors='coerce')
scrape_date                  = pd.to_datetime('2025-09-22')
airbnb['host_tenure_days']   = (scrape_date - airbnb['host_since']).dt.days
airbnb                       = airbnb.drop(columns=['host_since'])

# Standardize community area name to title case
airbnb = airbnb.rename(columns={'neighbourhood_cleansed': 'community_area'})
airbnb['community_area'] = airbnb['community_area'].str.title()

print(f"host_tenure_days nulls: {airbnb['host_tenure_days'].isna().sum()}")
print(airbnb['host_tenure_days'].describe().round(0))
```

### 1.3 Scope Filter — Short-Term Rentals Only

We identified 1,478 listings with zero reviews. Investigation showed these have a median minimum nights of 32 days and are long-term rentals listed on Airbnb, not the short-term rental market we are studying. Including them would add noise without contributing meaningful pricing signal. We remove them, leaving 6,346 listings.

We also drop null prices as it is our target variable and remove hotel/B&B property types, which operate under a different pricing model.

```python
# Drop null prices — target variable
airbnb = airbnb.dropna(subset=['price'])
print(f"After dropping null prices: {len(airbnb):,}")

# Remove zero-review listings (long-term rentals)
airbnb = airbnb[airbnb['number_of_reviews'] > 0]
print(f"After removing zero-review listings: {len(airbnb):,}")

# Remove hotel and B&B property types
remove_types = [
    'Room in hotel', 'Camper/RV', 'Room in boutique hotel',
    'Shared room in hotel', 'Room in bed and breakfast',
    'Private room in bed and breakfast', 'Farm stay', 'Private room in farm stay',
]
airbnb = airbnb[~airbnb['property_type'].isin(remove_types)]
print(f"After removing hotel/B&B types: {len(airbnb):,}")
```

### 1.4 Null Imputation

Three columns require imputation rather than row drops:

- `bedrooms` — filled from beds when null (same concept)
- `bathrooms` — filled from beds when null (reasonable assumption)
- `host_is_superhost` — 246 nulls (4% of rows). Null hosts show similar pricing (138 dollar median) and tenure (2,879 days) to known non-superhosts ($145, 2,903 days), so we fill with False rather than dropping rows
- `host_response_time` — nulls filled with category 'unknown'

Rows with null beds are dropped (very few) since no imputation basis exists.


```python
# Impute bedrooms and bathrooms from beds
airbnb['bedrooms']  = airbnb['bedrooms'].fillna(airbnb['beds'])
airbnb['bathrooms'] = airbnb['bathrooms'].fillna(airbnb['beds'])
airbnb = airbnb.dropna(subset=['beds'])
airbnb = airbnb.drop(columns=['beds'])

# Impute host_response_time unknowns
airbnb['host_response_time'] = (
    airbnb['host_response_time'].cat.add_categories('unknown').fillna('unknown')
)

# Impute host_is_superhost nulls with False
airbnb['host_is_superhost'] = airbnb['host_is_superhost'].fillna(False)

print(f"Final shape: {airbnb.shape}")
print(f"Remaining nulls:\n{airbnb.isnull().sum()[airbnb.isnull().sum() > 0]}")
```

```python
airbnb.to_csv('../data/cleaned/cleaned_listings.csv', index=False)
```

## Section 2 — Crime Data Cleaning

The Chicago Data Portal provides all reported incidents from the past year. We clean column names, fix data types, and filter to non-domestic incidents only. Domestic incidents are excluded because they reflect household conflict rather than public safety conditions that would affect a guest's experience.

**Source:** `Crimes_-_One_year_prior_to_present.csv` — Feb 2025 to Feb 2026


```python
crime_df = pd.read_csv('../data/raw/Crimes_-_One_year_prior_to_present.csv')
print(f"Raw shape: {crime_df.shape}")
print(crime_df.columns.tolist())
```

```python
# Rename columns and drop administrative fields
drop_cols = ['IUCR', ' IUCR', 'WARD', 'FBI CD']
crime_df = crime_df.drop(columns=drop_cols, errors='ignore')

crime_df = crime_df.rename(columns={
    'CASE#':                 'crime_id',
    'DATE  OF OCCURRENCE':   'date',
    ' PRIMARY DESCRIPTION':   'primary_type',
    ' SECONDARY DESCRIPTION': 'secondary_type',
    ' LOCATION DESCRIPTION':  'location_description',
    'ARREST':                'arrest',
    'DOMESTIC':              'domestic',
    'LATITUDE':              'latitude',
    'LONGITUDE':             'longitude',
    'X COORDINATE':          'x_coordinate',
    'Y COORDINATE':          'y_coordinate',
    'LOCATION':              'location',
    'BEAT':                  'beat',
    'BLOCK':                 'block'
})

# Drop columns not needed for spatial join
crime_df = crime_df.drop(columns=['block', 'x_coordinate', 'y_coordinate', 'location'], errors='ignore')

# Fix types
crime_df['date']     = pd.to_datetime(crime_df['date'], errors='coerce')
crime_df['domestic'] = crime_df['domestic'].map({'Y': True, 'N': False})
crime_df['arrest']   = crime_df['arrest'].map({'Y': True, 'N': False})

# Normalize text
crime_df['primary_type']   = crime_df['primary_type'].str.lower().str.strip()
crime_df['secondary_type'] = crime_df['secondary_type'].str.lower().str.strip()

# Drop nulls in critical fields
crime_df = crime_df.dropna(subset=['latitude', 'longitude', 'primary_type', 'domestic'])

# Keep non-domestic only
crime_df = crime_df[crime_df['domestic'] == False]

print(f"Cleaned shape: {crime_df.shape}")
print(f"Date range: {crime_df['date'].min().date()} to {crime_df['date'].max().date()}")
print(f"Remaining nulls: {crime_df.isnull().sum().sum()}")
```

```python
crime_df.to_csv('../data/cleaned/cleaned_crime.csv', index=False)
```

## Section 3 — Income Data Cleaning

Median household income estimates come from the Chicago Health Atlas, sourced from the 5-year ACS (2020–2024). The dataset covers all 77 Chicago community areas. We compute a margin of error percentage flag to identify areas where the ACS estimate is unreliable (MOE > 20% of the estimate).

**Source:** `Chicago Health Atlas Data Download - INC in Community areas.xlsx`


```python
income = pd.read_excel(
    '../data/raw/Chicago Health Atlas Data Download - INC in Community areas.xlsx',
    sheet_name='Community areas'
)
print(f"Raw shape: {income.shape}")
income.head(3)
```

```python
# Drop the 'Layer' column — duplicate with 'Community area'
income = income.drop(columns=['Layer'])

# Rename
income = income.rename(columns={
    'Name':              'community_area',
    'GEOID':             'area_number',
    'INC_2020-2024':     'median_income',
    'INC_2020-2024_moe': 'income_moe',
})

# Compute margin of error as % of estimate, flag unreliable areas (MOE > 20%)
income['income_moe_pct']    = (income['income_moe'] / income['median_income'] * 100).round(1)
income['income_unreliable'] = (income['income_moe_pct'] > 20).astype(int)

print(f"Shape: {income.shape}")
print(f"Unreliable areas (MOE > 20%): {income['income_unreliable'].sum()} of 77")
print(f"Income range: ${income['median_income'].min():,.0f} ({income.loc[income['median_income'].idxmin(), 'community_area']}) "
      f"— ${income['median_income'].max():,.0f} ({income.loc[income['median_income'].idxmax(), 'community_area']})")
```

```python
income.to_csv('../data/cleaned/cleaned_income.csv', index=False)
```

## Section 4 — Spatial Join: Crime Incidents to Community Areas

Each crime record has a GPS coordinate, but no community area label. We use the City of Chicago community area boundary polygons to assign every crime to its neighborhood via a spatial join. The result is an aggregated crime profile for each of the 77 community areas.

We compute five aggregated metrics:
- `total_crimes` — total non-domestic incidents
- `violent_crimes` — subset of explicitly violent types (battery, assault, robbery, homicide, etc.)
- `theft_count` — theft incidents specifically
- `arrest_rate` — proportion of incidents resulting in arrest
- `pct_violent` — violent crimes as a share of total


```python
# Load and prepare community area boundaries
boundaries_raw = pd.read_csv('../data/raw/Boundaries_-_Community_Areas_20260305.csv')

boundaries = boundaries_raw[['the_geom', 'AREA_NUMBE', 'COMMUNITY']].copy()
boundaries = boundaries.rename(columns={'AREA_NUMBE': 'area_number', 'COMMUNITY': 'community_area'})
boundaries['community_area'] = boundaries['community_area'].str.title()
boundaries['community_area'] = boundaries['community_area'].replace({
    'Mckinley Park': 'McKinley Park',
    'Ohare': "O'Hare"
})
boundaries['geometry'] = boundaries['the_geom'].apply(wkt.loads)
boundaries = gpd.GeoDataFrame(boundaries, geometry='geometry', crs='EPSG:4326')
boundaries = boundaries.drop(columns=['the_geom'])

# Verify name alignment with income
income_check = pd.read_csv('../data/cleaned/cleaned_income.csv')
unmatched = set(boundaries['community_area']) - set(income_check['community_area'])
assert len(unmatched) == 0, f"Name mismatch with income: {unmatched}"
print(f"Boundary names aligned with income — {len(boundaries)} areas")
```

```python
# Convert crime records to GeoDataFrame
crime = pd.read_csv('../data/cleaned/cleaned_crime.csv')
crime = crime.dropna(subset=['latitude', 'longitude'])

crime_gdf = gpd.GeoDataFrame(
    crime,
    geometry=gpd.points_from_xy(crime['longitude'], crime['latitude']),
    crs='EPSG:4326'
)
print(f"Crime GeoDataFrame: {crime_gdf.shape[0]:,} points, CRS: {crime_gdf.crs}")
```

```python
# Spatial join: assign each crime to its community area
crime_joined = gpd.sjoin(
    crime_gdf,
    boundaries[['area_number', 'community_area', 'geometry']],
    how='left',
    predicate='within'
)

outside = crime_joined['community_area'].isnull().sum()
print(f"Crimes outside all boundaries: {outside:,} ({outside/len(crime_joined)*100:.1f}%)")
```

```python
# Aggregate to community area level
violent_types = [
    'battery', 'assault', 'robbery', 'homicide',
    'criminal sexual assault', 'kidnapping', 'arson',
    'stalking', 'intimidation', 'offense involving children',
    'human trafficking', 'weapons violation'
]

crime_by_area = crime_joined.groupby('community_area').agg(
    area_number    = ('area_number',  'first'),
    total_crimes   = ('crime_id',     'count'),
    violent_crimes = ('primary_type', lambda x: x.isin(violent_types).sum()),
    theft_count    = ('primary_type', lambda x: (x == 'theft').sum()),
    arrest_rate    = ('arrest',       'mean'),
).reset_index()

crime_by_area['pct_violent'] = (crime_by_area['violent_crimes'] / crime_by_area['total_crimes']).round(4)

# Match 77 areas and fill areas with zero crimes
all_areas = boundaries[['area_number', 'community_area']]
crime_by_area = all_areas.merge(crime_by_area, on=['area_number', 'community_area'], how='left')
fill_cols = ['total_crimes', 'violent_crimes', 'theft_count', 'arrest_rate', 'pct_violent']
crime_by_area[fill_cols] = crime_by_area[fill_cols].fillna(0)
crime_by_area[['total_crimes', 'violent_crimes', 'theft_count']] = (
    crime_by_area[['total_crimes', 'violent_crimes', 'theft_count']].astype(int)
)

# Validation
assert len(crime_by_area) == 77
assert crime_by_area.isnull().sum().sum() == 0
print(f"crime_by_area: {crime_by_area.shape}")
print(crime_by_area.nlargest(5, 'total_crimes')[['community_area', 'total_crimes', 'violent_crimes', 'pct_violent']])
```

```python
crime_by_area.to_csv('../data/cleaned/crime_by_area.csv', index=False)
```

## Section 5 — Final Merge

We combine all three cleaned sources into one analysis-ready dataset. The join key is `community_area` and has been verified to align across all three tables before merging. Both merges are LEFT JOINs on listings to ensure zero rows are lost.

Two additional steps happen here:
- **Amenity count** — the raw amenities column is a JSON string list. We parse it and store the count as a numeric proxy for listing quality.
- **Host tenure nulls** — 182 listings have no host_since date. We fill with the median (2,903 days) rather than dropping, since these rows have valid prices and all other features. We use median over mean because tenure is right-skewed.


```python
listings   = pd.read_csv('../data/cleaned/cleaned_listings.csv')
income     = pd.read_csv('../data/cleaned/cleaned_income.csv')
crime_area = pd.read_csv('../data/cleaned/crime_by_area.csv')

print(f"Listings:   {listings.shape}")
print(f"Income:     {income.shape}")
print(f"Crime area: {crime_area.shape}")
```

```python
# Fix known casing edge cases
listings['community_area'] = listings['community_area'].replace({
    'Mckinley Park': 'McKinley Park',
    'Ohare': "O'Hare"
})

# Verify alignment
mismatch_income = set(listings['community_area']) - set(income['community_area'])
mismatch_crime  = set(listings['community_area']) - set(crime_area['community_area'])
assert len(mismatch_income) == 0, f"Income name mismatch: {mismatch_income}"
assert len(mismatch_crime)  == 0, f"Crime name mismatch: {mismatch_crime}"
print("All community area names aligned across tables")
```

```python
# Parse amenity count from JSON string
def count_amenities(x):
    try:
        return len(ast.literal_eval(x))
    except:
        return 0

listings['amenity_count'] = listings['amenities'].apply(count_amenities)
listings = listings.drop(columns=['amenities'])
print(f"Amenity count: min={listings['amenity_count'].min()}, "
      f"median={listings['amenity_count'].median()}, max={listings['amenity_count'].max()}")

# Drop columns that don't belong in the modeling dataset
listings = listings.drop(columns=['number_of_reviews', 'calculated_host_listings_count'])

# Fill host_tenure_days nulls with median (right-skewed distribution)
median_tenure = listings['host_tenure_days'].median()
listings['host_tenure_days'] = listings['host_tenure_days'].fillna(median_tenure)
print(f"Filled host_tenure_days nulls with median: {median_tenure:.0f} days")
```

```python
# Merge 1: listings + income
before = len(listings)
df = listings.merge(income, on='community_area', how='left')
assert len(df) == before, "Rows lost in income merge"

# Merge 2: + crime
before = len(df)
df = df.merge(
    crime_area[['community_area', 'total_crimes', 'violent_crimes',
                'theft_count', 'arrest_rate', 'pct_violent']],
    on='community_area', how='left'
)
assert len(df) == before, "Rows lost in crime merge"

print(f"Final shape: {df.shape}")
print(f"Community areas covered: {df['community_area'].nunique()}")
print(f"\nAreas with no listings: {set(income['community_area']) - set(df['community_area'])}")
```

```python
# Final validation
assert len(df) == 6162,                          "Unexpected row count"
assert df['price'].isnull().sum() == 0,          "Null prices"
assert df['median_income'].isnull().sum() == 0,  "Null income values"
assert df['total_crimes'].isnull().sum() == 0,   "Null crime values"
assert df['amenity_count'].isnull().sum() == 0,  "Null amenity counts"
assert df['host_tenure_days'].isnull().sum() == 0, "Null tenure days"
assert df['price'].min() > 0,                    "Zero or negative prices"

print("All assertions passed")
```

```python
df.to_csv('../data/cleaned/final_merged.csv', index=False)
```

## Variance Check for AirBnB Variables

**Threshold**: Variables where ≥90% of values are identical will be flagged as having insufficient variance.

```python
# Function to check variance for categorical variables
def check_categorical_variance(merged_df, var, threshold=0.90):
    """Check if a categorical variable has sufficient variance."""
    value_counts = merged_df[var].value_counts()
    if len(value_counts) == 0:
        return False, 1.0, "All missing"
    
    max_freq = value_counts.iloc[0]
    total = merged_df[var].notna().sum()
    max_pct = max_freq / total if total > 0 else 1.0
    
    most_common = value_counts.index[0]
    
    return max_pct < threshold, max_pct, most_common

# Check variance for categorical variables
variance_results_cat = []
for var in categorical_vars:
    sufficient_var, max_pct, most_common = check_categorical_variance(merged_df, var)
    variance_results_cat.append({
        'Variable': var,
        'Most_Common_Value': most_common,
        'Most_Common_Pct': f"{max_pct*100:.2f}%",
        'Sufficient_Variance': 'Yes' if sufficient_var else 'No'
    })

# Check variance for numerical variables
variance_results_num = []
for var in numerical_vars:
    if merged_df[var].notna().sum() > 0:
        std = merged_df[var].std()
        mean = merged_df[var].mean()
        cv = (std / mean * 100) if mean != 0 else 0  # Coefficient of variation
        
        # Check if essentially constant (very low CV or zero std)
        sufficient_var = std > 0.001 and cv > 0.1
        
        variance_results_num.append({
            'Variable': var,
            'Std_Dev': f"{std:.4f}",
            'Mean': f"{mean:.4f}",
            'Coef_Variation': f"{cv:.2f}%",
            'Sufficient_Variance': 'Yes' if sufficient_var else 'No'
        })

# TABLE 2: Variance Check Results

display(pd.DataFrame(variance_results_cat))
display(pd.DataFrame(variance_results_num))
```

# Section 7 - Feature Engineering

Before modeling, we engineer one new feature from the existing data:

Before building the feature list, we check the correlation between `accommodates` 
and `bedrooms`. If r > 0.8, we should only keep one to avoid multicollinearity. 
We keep `accommodates` because it has a stronger correlation with price and is 
more directly actionable for a host — a host can add sleeping capacity more 
easily than adding bedrooms.

```python
print(merged_df[['accommodates', 'bedrooms']].corr())
```

**V2** adds two neighborhood variables selected from EDA Parts 2 and 3:

- **theft_count** chosen over `total_crimes` and `violent_crimes` because it has 
the highest correlation with price (r=0.16) and avoids multicollinearity 
(`total_crimes` and `violent_crimes` are correlated at r=0.96).

```python
# PREP: Cap price at 99th percentile to remove extreme outliers that would skew model training. 

# Cap price outliers based on EDA distribution
price_cap = merged_df['price'].quantile(0.99)
df = merged_df[merged_df['price'] <= price_cap].copy()

# One-hot encode room_type and property_type. drop_first=True avoids the dummy variable trap
df = pd.get_dummies(df, columns=['room_type', 'property_type'], drop_first=True)

# Update feature lists with dummy columns
room_dummies = [c for c in df.columns if c.startswith('room_type_')]
property_dummies = [c for c in df.columns if c.startswith('property_type_')]

features_v1 = [
    'accommodates', 'bathrooms', 'minimum_nights',
    'review_scores_cleanliness', 'review_scores_location', 'review_scores_rating',
    'host_is_superhost', 'instant_bookable',
    'amenity_count', 'host_tenure_days',
    'is_entire_home'
] + room_dummies + property_dummies

features_v2 = features_v1 + ['theft_count', 'median_income']
```

# Section 8 - Modeling

### Linear Regression: V1 (Airbnb Variables Only)

V1 uses only listing-level features to establish a baseline — how well can 
we predict price from what a host directly controls?

```python
# ── MODEL V1 ──────────────────────────────────────────────────────────────────
df_v1 = df[features_v1 + ['price']].dropna()
X1 = df_v1.drop(columns=['price'])
y1 = df_v1['price']

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42
)

scaler1 = StandardScaler()
X1_train_s = scaler1.fit_transform(X1_train)
X1_test_s  = scaler1.transform(X1_test)

lr1 = LinearRegression()
lr1.fit(X1_train_s, y1_train)
y1_pred = lr1.predict(X1_test_s)

print("\n=== LINEAR MODEL V1: Airbnb Variables Only ===")
print(f"R²:   {r2_score(y1_test, y1_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y1_test, y1_pred)):.2f}")
print(f"MAE:  {mean_absolute_error(y1_test, y1_pred):.2f}")
```

### Linear Regression: V2 (+ Crime & Income)

V2 adds neighborhood context to test whether knowing a listing's crime 
environment and median area income improves price prediction.

```python
# ── MODEL V2 ──────────────────────────────────────────────────────────────────
df_v2 = df[features_v2 + ['price']].dropna()
X2 = df_v2.drop(columns=['price'])
y2 = df_v2['price']

X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2, test_size=0.2, random_state=42
)

scaler2 = StandardScaler()
X2_train_s = scaler2.fit_transform(X2_train)
X2_test_s  = scaler2.transform(X2_test)

lr2 = LinearRegression()
lr2.fit(X2_train_s, y2_train)
y2_pred = lr2.predict(X2_test_s)

print("\n=== LINEAR MODEL V2: Airbnb + Crime + Income ===")
print(f"R²:   {r2_score(y2_test, y2_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y2_test, y2_pred)):.2f}")
print(f"MAE:  {mean_absolute_error(y2_test, y2_pred):.2f}")
```

means accuracy is a reliable metric here.

```python
# ── CREATE TARGET VARIABLE 
median_price = merged_df['price'].median()
df['is_high_price'] = (df['price'] > median_price).astype(int)

print(f"Median price threshold: ${median_price:.2f}")
print(df['is_high_price'].value_counts())
print(f"Class balance: {df['is_high_price'].mean():.1%} high price listings")
```

### Logistic Regression: V1 (Airbnb Variables Only)

Using the same features as Linear V1, we now predict the binary outcome 
(high vs. low price) rather than the exact price.

```python
df_log1 = df[features_v1 + ['is_high_price']].dropna()
X_log1 = df_log1.drop(columns=['is_high_price'])
y_log1 = df_log1['is_high_price']

X_log1_train, X_log1_test, y_log1_train, y_log1_test = train_test_split(
    X_log1, y_log1, test_size=0.2, random_state=42
)

scaler_log1 = StandardScaler()
X_log1_train_s = scaler_log1.fit_transform(X_log1_train)
X_log1_test_s  = scaler_log1.transform(X_log1_test)

log1 = LogisticRegression(max_iter=1000, random_state=42)
log1.fit(X_log1_train_s, y_log1_train)
y_log1_pred = log1.predict(X_log1_test_s)

print("\n=== LOGISTIC V1: Airbnb Variables Only ===")
print(f"Accuracy:  {accuracy_score(y_log1_test, y_log1_pred):.4f}")
print(f"Precision: {precision_score(y_log1_test, y_log1_pred):.4f}")
print(f"Recall:    {recall_score(y_log1_test, y_log1_pred):.4f}")
print(f"F1 Score:  {f1_score(y_log1_test, y_log1_pred):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_log1_test, log1.predict_proba(X_log1_test_s)[:,1]):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_log1_test, y_log1_pred))
print("\nClassification Report:")
print(classification_report(y_log1_test, y_log1_pred))
```

### Logistic Regression: V2 (+ Crime & Income)

Adding crime and income variables — do they help classify listings more 
accurately beyond listing characteristics?

```python
df_log2 = df[features_v2 + ['is_high_price']].dropna()
X_log2 = df_log2.drop(columns=['is_high_price'])
y_log2 = df_log2['is_high_price']

X_log2_train, X_log2_test, y_log2_train, y_log2_test = train_test_split(
    X_log2, y_log2, test_size=0.2, random_state=42
)

scaler_log2 = StandardScaler()
X_log2_train_s = scaler_log2.fit_transform(X_log2_train)
X_log2_test_s  = scaler_log2.transform(X_log2_test)

log2 = LogisticRegression(max_iter=1000, random_state=42)
log2.fit(X_log2_train_s, y_log2_train)
y_log2_pred = log2.predict(X_log2_test_s)

print("\n=== LOGISTIC V2: Airbnb + Crime + Income ===")
print(f"Accuracy:  {accuracy_score(y_log2_test, y_log2_pred):.4f}")
print(f"Precision: {precision_score(y_log2_test, y_log2_pred):.4f}")
print(f"Recall:    {recall_score(y_log2_test, y_log2_pred):.4f}")
print(f"F1 Score:  {f1_score(y_log2_test, y_log2_pred):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_log2_test, log2.predict_proba(X_log2_test_s)[:,1]):.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_log2_test, y_log2_pred))
print("\nClassification Report:")
print(classification_report(y_log2_test, y_log2_pred))
```

We built four models to answer our research question:
*"What characteristics define a high-revenue Chicago Airbnb listing, and how should a new host price and position their listing to maximize earnings?"*

### Linear Regression (Predicting Exact Nightly Price)
- V1 (listing features only): R² = 0.4755, RMSE = 112.02/night, MAE = 68.90/night
- V2 (+ neighborhood crime and income): R² = 0.5303, RMSE = 106.01/night, MAE = 63.94/night
- Adding neighborhood crime and income data improved R² by ~5.5%, reducing prediction error by ~$6/night

### Logistic Regression (Predicting Above/Below Median Price of $143/night)
- V1 (listing features only): Accuracy = 78.95%, AUC = 0.876
- V2 (+ neighborhood crime and income): Accuracy = 81.57%, AUC = 0.900
- V2 correctly classifies 4 out of 5 listings as high or low price
- Adding crime and income data improved accuracy by 2.6% and AUC by 0.024

### Key Drivers of High Airbnb Pricing in Chicago
1. **Physical capacity** — accommodates and bathrooms are the strongest predictors across both model types
2. **Property type** — entire homes command significant premiums over private or shared rooms
3. **Neighborhood context** — crime and income variables meaningfully improved both models, confirming that location beyond community area name adds predictive value
4. **Amenities** — amenity_count is consistently associated with higher prices
5. **The "tourist area" effect** — higher theft_count areas (downtown/Near North Side) actually command higher prices, suggesting tourism density outweighs crime concerns for guests

### Limitations
- R² of ~0.53 means roughly half of price variance is still unexplained — seasonality, listing photos, and description quality are likely missing factors
- Linear regression assumes linear relationships, which may not fully capture neighborhood pricing effects
- Data is a single snapshot (September 2025) — seasonal pricing patterns are not captured
- Model is capped at the 99th percentile and does not generalize to luxury or hotel-style listings
- Analysis is specific to Chicago's dense urban market and may not generalize to other cities
- Future work: incorporate calendar availability and seasonal pricing data
